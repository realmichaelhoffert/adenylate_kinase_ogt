```python
from ete3 import NCBITaxa
ncbi = NCBITaxa()
```

```python
taxids = [6183]
lineages = ncbi.get_lineage(taxids[0]) # Returns list of parent TaxIDs
names = ncbi.get_taxid_translator(lineages) # Translates IDs to names
full_string = "; ".join([names[taxid] for taxid in lineages])
print(full_string)
```
